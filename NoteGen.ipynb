# Code for Generating Musical Audio Signals

```python
import re

TUNING_A4_440 = 440
SEMITONES_PER_OCTAVE = 12

SEMITONES_FROM_A = {
    'A': 0,
    'A#': 1,
    'Bb': 1,
    'B': 2,
    'C': -9,
    'C#': -8,
    'Db': -8,
    'D': -7,
    'D#': -6,
    'Eb': -6,
    'E': -5,
    'F': -4,
    'F#': -3,
    'Gb': -3,
    'G': -2,
    'G#': -1,
    'Ab': -1
}


def note_freq(note_name, transpose=0, a4_freq=TUNING_A4_440):
    """ Determine frequency corresponding to a musical note
    
    Args:
        x: the name of the note in the form 'Eb4', for example
        transpose: number of semitones to transpose notes
        a4_freq: the frequency, in Hz, corresponding to the note A4
    Returns:
        The frequency, in Hz, of the note
        
    """
    result = re.match(r'([A-G][b#]?)([0-8])', note_name)    
    note = result.group(1)
    octave = int(result.group(2))
    
    freq = a4_freq * (2 ** (octave - 4 + ((SEMITONES_FROM_A[note] + transpose) / float(SEMITONES_PER_OCTAVE))))
    
    return freq

def generate_waveform(note_sequence, fs, bpm, transpose=0):
    """ Generate an audio waveform from musical notes
    
    Args:
        note_sequence: an array of 2-item tuples, where the
                       first item is an array of note names,
                       and the second item is the duration
                       for those notes, in beats
        fs: the sampling frequency for the signal
        bpm: the number of beats per minute for the notes
        transpose: number of semitones to transpose notes
    Returns:
        A numpy array containing the audio signal
    
    """
    samples_per_beat = 60 * fs / float(bpm)
    signal = []
    
    # Iterate through each note in the sequence
    for notes,duration in note_sequence:
        
        duration_samples = int(duration * samples_per_beat)
        temp = zeros(duration_samples)
        
        for note in notes:
            # Compute the frequency for this note
            freq_hz = note_freq(note, transpose)
            freq_normalized = freq_hz / float(fs)
        
            # Create a tone
            temp += sin(2 * pi * freq_normalized * arange(duration_samples))
        
        signal = hstack([signal, temp])
        
    # Normalize the signal
    signal /= max(signal)
        
    return signal
```
